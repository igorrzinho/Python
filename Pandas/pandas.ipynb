### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha
display(vendas_df.loc[5]) #  vai trazer a linha 5
```

```python
# pegar varias linha
display(vendas_df.loc[5:10]) # da linha 5 ate a linha 10
```

```python
# pegar todas as linhas que correspondem a uma condição

display(vendas_df.loc[vendas_df['ID Loja']=='Norte Shopping'])
```

```python
# pegar varias linhas e colunas usando .loc
vendas_norte_shopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norte_shopping)
```

### Adicionar 1 coluna

```python
# criar uma coluna comissao que vai ser 5% do valor de vendsa
vendas_df['Comissao'] = vendas_df['Valor Final']* 0.05
display(vendas_df)
```

```python
vendas_df.loc[:,'Imposto'] = 1
display(vendas_df)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
 vendas_dez_df = pd.read_excel('/content/Vendas - Dez.xlsx')
 display(vendas_dez_df)
```

```python
vendas_df = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df)
```

### Excluir linhas e colunas

```python
# excluir uma linha
vendas_df = vendas_df.drop('Imposto', axis=1)
```

```python
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
# axis = 0 linha
# axis = 1 coluna
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# deletar linhas vazias
vendas_df = vendas_df.dropna(how='all',axis=1)
```

```python
# deletar linhas que possuam um valor vazio
vendas_df = vendas_df.dropna()
```

```python
# preencher valores vazios
# com media da coluna
vendas_df['Comissao'] = vendas_df['Comissao'].fillna(vendas_df['Comissao'].mean())
```

```python
# preencher com o ultimo valor
vendas_df = vendas_df.ffill()
```

### Calcular Indicadores
    - Value Counts

```python
idlojas = vendas_df['ID Loja'].value_counts()
display(idlojas)
```

    - Groupby

```python
faturamento_produto = vendas_df.groupby('Produto')['Valor Final'].sum()
display(faturamento_produto)
```
